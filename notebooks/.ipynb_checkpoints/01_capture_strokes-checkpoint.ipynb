# Capture Raw Strokes (Notebook Canvas)

This notebook provides a simple canvas (mouse/touch) to capture raw stroke paths per letter and save them as JSON files for the stroke-based Conv1D+BiGRU pipeline.

- Output format matches the loader: `{ "points": [[x,y,t,stroke_id], ...], "user_id": "local", "session_id": "notebook" }`
- Saved under `data/strokes/<split>/<letter>/<timestamp>.json`.

```python
# Install if needed (uncomment in a new environment)
# %pip install ipywidgets ipycanvas
import os, json, time
from pathlib import Path
import ipywidgets as W
from ipycanvas import Canvas, hold_canvas

LETTERS = ['א','ב','ג','ד','ה','ו','ז','ח','ט','י','ך','כ','ל','ם','מ','ן','נ','ס','ע','ף','פ','ץ','צ','ק','ר','ש','ת']
BASE = Path('data/strokes')
(BASE / 'train').mkdir(parents=True, exist_ok=True)
(BASE / 'val').mkdir(parents=True, exist_ok=True)
(BASE / 'test').mkdir(parents=True, exist_ok=True)
```

```python
# UI Elements
split_dd = W.Dropdown(options=['train','val','test'], value='train', description='Split')
letter_dd = W.Dropdown(options=LETTERS, value='א', description='Letter')
clear_btn = W.Button(description='Clear')
undo_btn = W.Button(description='Undo')
save_btn = W.Button(description='Save')
status = W.HTML(value='Ready.')

# Canvas setup
SIZE = 320
canvas = Canvas(width=SIZE, height=SIZE)
canvas.layout.border = '1px solid #aaa'

# State: list of strokes, each stroke is list of dicts {x,y,t}
strokes = []
curr = []
drawing = False

def redraw():
    with hold_canvas(canvas):
        canvas.clear()
        canvas.fill_style = 'white'
        canvas.fill_rect(0, 0, SIZE, SIZE)
        canvas.stroke_style = 'black'
        canvas.line_width = 3
        for s in strokes + ([curr] if curr else []):
            if len(s) < 1:
                continue
            canvas.begin_path()
            canvas.move_to(s[0]['x'], s[0]['y'])
            for p in s[1:]:
                canvas.line_to(p['x'], p['y'])
            canvas.stroke()

def on_mouse_down(x, y):
    global drawing, curr
    drawing = True
    curr = [{ 'x': x, 'y': y, 't': time.time() }]
    redraw()

def on_mouse_move(x, y):
    if not drawing:
        return
    curr.append({ 'x': x, 'y': y, 't': time.time() })
    redraw()

def on_mouse_up(x, y):
    global drawing, curr
    if not drawing:
        return
    drawing = False
    if len(curr) > 1:
        strokes.append(curr)
    curr = []
    redraw()

canvas.on_mouse_down(on_mouse_down)
canvas.on_mouse_move(on_mouse_move)
canvas.on_mouse_up(on_mouse_up)

def on_clear(_):
    global strokes, curr
    strokes = []
    curr = []
    redraw()

def on_undo(_):
    if strokes:
        strokes.pop()
        redraw()

def on_save(_):
    # Flatten to points [[x,y,t,stroke_id], ...]
    pts = []
    for sid, s in enumerate(strokes):
        for p in s:
            pts.append([float(p['x']), float(p['y']), float(p['t']), float(sid)])
    if not pts:
        status.value = "<span style='color:#c00'>No strokes to save.</span>"
        return
    split = split_dd.value
    letter = letter_dd.value
    out_dir = BASE / split / letter
    out_dir.mkdir(parents=True, exist_ok=True)
    ts = int(time.time() * 1000)
    out_path = out_dir / f'{ts}.json'
    with open(out_path, 'w', encoding='utf-8') as f:
        json.dump({ 'points': pts, 'user_id': 'local', 'session_id': 'notebook' }, f, ensure_ascii=False)
    status.value = f'Saved to <code>{out_path}</code>'

clear_btn.on_click(on_clear)
undo_btn.on_click(on_undo)
save_btn.on_click(on_save)

W.VBox([
    W.HBox([split_dd, letter_dd, clear_btn, undo_btn, save_btn]),
    canvas,
    status,
])
```

Tip: draw ~10–20 samples per letter to start; then run the stroke model training:

`PYTHONPATH=src python -m stroke_recognizer.train --data-root data --epochs 40 --batch-size 64 --n-points 96`
